```python
import ast
import re
def safe_eval_hashtags(x):
    if pd.notnull(x):
        try:
            # Remove non-printable characters and parse
            cleaned = re.sub(r'[^\x20-\x7E]', '', x)
            hashtags_dict = ast.literal_eval(cleaned)
            return len(hashtags_dict)
        except (ValueError, SyntaxError):
            # Handle invalid dictionary strings
            return 0
    return 0

df['Total_Hashtags'] = df['Hashtags_(dictionary)'].apply(safe_eval_hashtags)
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values
# Option 1: Fill missing numeric columns with 0 or the mean
numeric_columns = ['Followers', 'Average Likes', 'Average Comments',
                   'Average Engagement Rate', 'Engagement Rate (60 Days)',
                   'Posting frequency (last 60 days)', 'Posts (image)',
                   'Posts (videos)', 'Posts (carousel)', 'Average Hashtags / Post (60 Days)']

#df[numeric_columns] = df[numeric_columns].fillna(0)  # Replace with 0 or mean if appropriate

# Option 2: Drop rows with missing essential data (like Username)
#df = df.dropna(subset=['Username'])  # Remove rows without usernames
```

```python
# Convert numeric columns to float/int if not already
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert 'Hashtags (dictionary)' to string if needed
df['Hashtags (dictionary)'] = df['Hashtags (dictionary)'].astype(str)
```

```python
import ast
import re

# Clean and parse the dictionary
def parse_hashtags(x):
    if pd.isnull(x): return {}
    try:
        cleaned = re.sub(r'[^\x20-\x7E]', '', x)  # Remove non-printable characters
        return ast.literal_eval(cleaned)
    except (ValueError, SyntaxError):
        return {}

df['Parsed_Hashtags'] = df['Hashtags (dictionary)'].apply(parse_hashtags)
```

```python
# Total hashtags (sum of counts)
df['Total_Hashtags'] = df['Parsed_Hashtags'].apply(lambda x: sum(x.values()))

# Unique hashtags (number of distinct hashtags)
df['Unique_Hashtags'] = df['Parsed_Hashtags'].apply(len)
```

```python
# Visualize numeric columns to detect outliers
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df[numeric_columns])
plt.show()

# Example: Remove rows where 'Average Likes' is unusually high
q1 = df['Average Likes'].quantile(0.25)
q3 = df['Average Likes'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

df = df[(df['Average Likes'] >= lower_bound) & (df['Average Likes'] <= upper_bound)]
```

```python
# Calculate IQR bounds
q1 = df['Average Likes'].quantile(0.25)
q3 = df['Average Likes'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Add a column to flag outliers
df['Outlier_Avg_Likes'] = (df['Average Likes'] < lower_bound) | (df['Average Likes'] > upper_bound)
```
