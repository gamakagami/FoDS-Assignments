Apply functions to a dataframe df.apply() and df.applymap()

.apply()

```python
def rating_categorize(ratings):
    if pd.isna(ratings) == "NaN":
        return 'NaN'
    elif ratings < 7:
        return 'Low Rating'
    elif ratings < 8:
        return 'Average Rating'
    elif ratings < 9:
        return 'Good Rating'
    else:
        return 'Excellent Rating'

df_merged['Rating Group'] = df_merged['rating'].apply(rating_categorize)
df_merged
```

.applymap()

```python
def rating_categorize(ratings):
    if pd.isna(ratings):
        return 'NaN'
    elif ratings < 7:
        return 'Low Rating'
    elif ratings < 8:
        return 'Average Rating'
    elif ratings < 9:
        return 'Good Rating'
    else:
        return 'Excellent Rating'

df_categorized = df_merged.copy()

df_categorized['rating'] = df_categorized[['rating']].applymap(rating_categorize)

print(df_categorized)
```

Find the data which satisfy specific condition using df.query()

Perform grouping and aggregating operations using df.groupby() and df.agg().

groupby()

```python
grouped = df_merged.groupby('user_id')

average_rating = grouped['rating'].mean()
print(average_rating)
```

agg()

```python
agg_results = df_merged.groupby('user_id')['rating'].agg(['min', 'max', 'sum']).reset_index()

print(agg_results)
```

 imputing missing values using the SimpleImputer Class in sklearn

```python
from sklearn.impute import SimpleImputer

df = pd.DataFrame(df1)
print("Original DataFrame:")
print(df)

imputer_constant = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=999)

df_imputed_constant = df.copy()

columns_to_impute = ['user_id', 'rating']
df_imputed_constant[columns_to_impute] = imputer_constant.fit_transform(df_imputed_constant[columns_to_impute])

print("\nDataFrame after imputing missing values with constant (999):")
print(df_imputed_constant)
```
